```python
# ANNEAL
# 1) Generate an initial structure by running "python make_NP.py" in the terminal.
# 2) Visualize this initial structure
atoms = read("structures/atoms0.xyz")
viewer = WeasWidget()
viewer.avr.model_style = 0 # ball mode
viewer.from_ase(atoms)
viewer
```
